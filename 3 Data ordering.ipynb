## Группировка строк по времени

```python
import pandas as pd
import numpy as np


time_index = pd.date_range('06/06/2017', periods=1000000,freq='30s')

time_index
```

```python
dataframe = pd.DataFrame(index= time_index)

dataframe['Sale_Amount'] = np.random.randint(1,10,1000000)

dataframe
```

```python
dataframe.resample('W').sum()
```

```python
dataframe.resample('2W').mean()
```

```python
dataframe.resample('M').count()
```

```python
dataframe.resample('M', label = 'left').count()
```

## Слияние фреймов данных

```python
employee_data = {'employee_id':['1','2','3','4'],
                'name' : ['Amy Jones', 'Allen Keys', "Alise Bees", 'Tim Horton']}

dataframe_employees = pd.DataFrame(employee_data, columns=['employee_id', 'name'])

dataframe_employees
```

```python
sales_data = {'employee_id': ['3','4','5','6'],
             'total_sales': [2341,1245,866,2111]}
dataframe_sales = pd.DataFrame(sales_data,columns = ['employee_id', 'total_sales'])

dataframe_sales
```

```python
pd.merge(dataframe_employees,dataframe_sales, on = 'employee_id')
```

```python
pd.merge(dataframe_employees,dataframe_sales, on = 'employee_id', how = 'left')
```

```python
pd.merge(dataframe_employees,dataframe_sales, on = 'employee_id', how = 'outer')
```

```python
dataframe_sales.rename(columns = {'employee_id':'id'},inplace=True)
dataframe_sales
```

```python
pd.merge(dataframe_employees,dataframe_sales, left_on = 'employee_id', right_on = 'id')
```
