```python
fig, ax = plt.subplots()

x = [-3,-2,-1,0,1,2,3]
f_arrpow2 = lambda arr: [item**2 for item in arr]
y=f_arrpow2(x)

ax.plot(x,y) #== plt.plot(x,y)
y
```

```python
fig, ax = plt.subplots()
x = [-3,-2,-1,0,1,2,3]
f_arrpow2 = lambda arr: [item**2 for item in arr]
y=f_arrpow2(x)
f_arrabs = lambda arr: [abs(item) for item in arr]
x_abs = f_arrabs(x)
ax.plot(x,y)
ax.plot(x, x_abs)
```

```python
fig, axes = plt.subplots(2,2)
x = [-3,-2,-1,0,1,2,3]
f_arrpow2 = lambda arr: [item**2 for item in arr]
y=f_arrpow2(x)

axes[1,0].plot(x,y)
```

## Цвет линии и стили

```python
import numpy as np

fig, ax = plt.subplots()
x = np.array([-3,-2,-1,0,1,2,3])
y = x**2

ax.plot(x,y)

ax.plot(x, np.sin(x), color=(1.0,0.2,0.3), linestyle='-')

ax.plot(x, x+5, color='blue', linestyle = '--')

ax.plot(x, x+3, color='g', linestyle = ':')

ax.plot(x, np.cos(x), color='0.75', linestyle='-.')

ax.plot(x,x,color='#FFDD44', linestyle='--')
```

```python
fig,ax = plt.subplots()

x = np.array([-3,-2,-1,0,1,2,3])
y=x**2

ax.plot(x,y,marker='o')

ax.plot(0.5*x,y,marker='^')

ax.plot(2*x,y,marker='*')
```
